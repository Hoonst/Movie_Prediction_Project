```python
driver = webdriver.Chrome('chromedriver')
```

```python
def bring_my_table(movie):
    movie_dic[movie] = []
    #Input Movie Name
    #movie 파마리터를 가져오는데 이것을 반복문으로 영화 목록으로 설정한다.
    element_id = driver.find_element_by_name('sMovName')
    element_id.send_keys(movie)
    element_id.submit()
    time.sleep(2.5)
    try:
        driver.find_element_by_partial_link_text(movie[3]).click()
    except:
        try:
            driver.find_element_by_partial_link_text(movie[2]).click()
        except:
            try:
                driver.find_element_by_partial_link_text(movie[1]).click()
            except:                        
                print("something wrong here", movie)
                
    
    html= driver.page_source
    soup= BeautifulSoup(html, 'html.parser')
    
    #running time이 a[5].text에 존재하는데 ex) 154분 23초 이런식으로 되어 있어서
    # 숫자만 가져오고 그것을 .으로 합쳐버린다.
    a = soup.select('dd > ul > li')
    running = a[5].text
    pattern = '\d+'
    regex = re.compile(pattern)
    match = regex.findall(running)

    timeo = ".".join(match)
    timeo = float(timeo)

    actors = soup.select("td > div > a")

    #영화 배우 목록 Get
    for i in actors:
        movie_dic[movie].append(i.text)
    
    movie_dic[movie].append(timeo)
    
    
    
    driver.find_element_by_link_text('통계정보').click()
    
    #Explicit wait과 같은 함수로 페이지가 로드되면 진행되게 하려 했는데 실패하여 가장 큰 숫자인 30으로 진행
    time.sleep(30)
    #Achieving HTML
    html= driver.page_source
    soup= BeautifulSoup(html, 'html.parser')
    #Generating Table
    table = soup.find("caption", text="박스오피스").find_parent("table")
    
    columns= table.select('thead > tr > th')
    columns = columns[0:]
    columnlist=[]
    for column in columns:
        columnlist.append(column.text)
    df=pd.DataFrame(columns=columnlist)
    
    contents=table.select('tbody > tr')
    
    dfcontent=[]
    alldfcontents=[]

    for content in contents:
        tds=content.find_all("td")
        for td in tds:
            dfcontent.append(td.text)
        alldfcontents.append(dfcontent)
        dfcontent=[]
        
    #Final table DataFrame    
    df = pd.DataFrame(columns = columnlist, data = alldfcontents)
    
    return df
```

```python
def make_table(test):
    test = test.T
    
    screen = test.loc['스크린수'].T
    screen.index = [str(i) + 'screen' for i in range(0,11)]
    shown = test.loc['상영횟수'].T
    shown.index = [str(i) + 'shown' for i in range(0,11)]
    goers = test.loc['관객수'].T
    goers.index = [str(i) + 'goers' for i in range(0,11)]
    rank = test.loc['순위'].T
    rank.index = [str(i) + 'rank' for i in range(0,11)]
    
    total = pd.concat([screen,shown,goers,rank])
    table = pd.DataFrame(total).T
    
    
    return table
```

```python
num = 1
for i in movie_list:
    driver.get('http://www.kobis.or.kr/kobis/business/mast/mvie/searchMovieList.do')
    info = bring_my_table(i)
    print("Done bring_my_table", i)
    table = make_table(info)
    table.index = [num]
    base_table = base_table.append(table)
    num += 1
```
